```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('datasets/credit card taiwan svm algorithm.csv')
```

```python
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# handling missing values
null_handle = df.AGE.mean()
null_handle
```

```python
df.AGE = df.AGE.fillna(null_handle)
```

```python
df.isnull().sum()
```

```python
# spliting the dataset
x = df.drop(['default.payment.next.month'], axis=1)
y = df['default.payment.next.month']
```

```python
# x
y
```

```python
# Model fit
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = .3, random_state = 5)
```

```python
xtrain
```

```python
model = SVC()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

# lets apply random forests

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfcl_model = RandomForestClassifier()
```

```python
rfcl_model.fit(xtrain,ytrain)
rfcl_model.score(xtest,ytest)
```

# Lets apply decision tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier()
```

```python
dt_model.fit(xtrain,ytrain)
dt_model.score(xtest,ytest)
```
